```python
from src.definitions import EXTERNAL_DATA_FOLDER
from src.util.kaggle import download_and_unzip


ds_path = download_and_unzip(
    "sciencelabwork",
    "hyperspectral-image-sensing-dataset-ground-truth",
    EXTERNAL_DATA_FOLDER,
)
```

```python
import scipy.io

x = scipy.io.loadmat(ds_path / "Indian_pines_corrected.mat")
```

```python
x.keys()
```

```python
image = x["indian_pines_corrected"]

image.shape
```

```python
from matplotlib import pyplot as plt
import numpy as np


rgb_image = np.stack([image[:, :, 29], image[:, :, 19], image[:, :, 9]], axis=-1)

rgb_image = (rgb_image - rgb_image.min()) / (rgb_image.max() - rgb_image.min())

plt.imshow(rgb_image)
plt.title("False Color Composite of Indian Pines")
plt.show()
```

```python
gt_data = scipy.io.loadmat(ds_path / "Indian_pines_gt.mat")
ground_truth = gt_data["indian_pines_gt"]

plt.imshow(ground_truth, cmap="jet")
plt.colorbar()
plt.title("Ground Truth Labels")
plt.show()
```
